```python
import newspaper
from urllib.parse import urlparse, urljoin
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://malpedia.caad.fkie.fraunhofer.de'
url_mal_family = 'https://malpedia.caad.fkie.fraunhofer.de/families'
url_actors = 'https://malpedia.caad.fkie.fraunhofer.de/actors'
```

```python
def get_data_for_actor(url):
    mal_families = []
    references = []
    soup = BeautifulSoup(requests.get(url).content, "html.parser")
    for node in soup.findAll('span'):
        href = node.attrs.get("data-href")
        if href is not None:
            mal_families.append(href.split('/')[2])
            #print(href)
            
    for node in soup.findAll('tr'):
        href = node.attrs.get("data-href")
        if href is not None and href!="DEST_URL":
            references.append(href)
            #print(href)
    return mal_families, references
```

```python
malware_families = []
soup = BeautifulSoup(requests.get(url_mal_family).content, "html.parser")
for a_tag in soup.findAll("tr"):
    href = a_tag.attrs.get("data-href")
    print(href)
    if href is not None:
        malware_families.append(href.split('/')[2])
print(malware_families)
```

```python
actors = {}
soup = BeautifulSoup(requests.get(url_actors).content, "html.parser")
for a_tag in soup.findAll("tr"):
    href = a_tag.attrs.get("data-href")
    if href is not None:
        actor = href.split('/')[2]
        mal_families, references = get_data_for_actor(base_url+href)
        actors[actor] = {}
print(actors)
```
